## Bài 1. Cho hàm số:
$$
f(x)= x^2sin(x)e^{-x} \quad\quad [0 \leq x \leq 10]
$$
## Yêu Cầu
1. Trên môi trường Code:  
    1.1 Tạo biến $x \in [0,10]$  
    1.2 Tạo hàm $f(x)$  
    1.3 Vẽ biểu đồ tô đầy $x- f(x)$  
2. Trên môi trường Markdown: Gõ lệnh để tạo văn bản giống như đề bài 

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
a = 0; b = 10; n = 100
x = np.linspace(a, b, n)
y = x**2 *np.sin(x)*np.exp(-x)
plt.figure(figsize=(6,2))
plt.plot(x, y)
plt.show()
```

## Bài 2. Cho hàm số:
$$f(x)=
\begin{cases}
    \dfrac {\sqrt{2x^4+5}}{e^{sinx-1}} \quad\quad [0 \leq x \leq 2]\\
    cos(x) \times x^3 \quad\quad [ 0 \leq x \leq 5]\\
\end{cases}
$$
## Yêu Cầu
1. Trên môi trường Code:  
    1.1 Tạo biến $x \in [-5,0][0.5]$  
    1.2 Tạo hàm $f(x)$  
    1.3 Vẽ biểu đồ tô đầy $x- f(x)$  
2. Trên môi trường Markdown: Gõ lệnh để tạo văn bản giống như đề bài 

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
a = -5; b = 0; c= 5; n = 100
x1 = np.linspace(a, b, n)
y1 = (np.sqrt(2*x1**4 +5))/(np.exp(np.sin(x1-1)))
x2 = np.linspace(b, c, n)
y2 = np.cos(x2)* x2**3
x = np.r_[x1, x2];  y = np.r_[y1, y2]
plt.figure(figsize=(4,1))
plt.plot(x, y)
plt.show()
```

## Bài 3. Cho hàm số:
$$M(x)=
\begin{cases}
    7 \quad\quad [0 \leq x \leq 2]\\
    2x^2 - 5x -2 \quad\quad [ 2 \leq x \leq 4]\\
    -3x +2 \quad\quad [4 \leq x \leq 5]
\end{cases}
$$  
## Yêu Cầu
1. Trên môi trường Code:  
    1.1 Tạo biến $x \in [0,2][2,4][4,5]$  
    1.2 Tạo hàm $M(x)$  
    1.3 Vẽ biểu đồ tô đầy $x- M(x)$  
2. Trên môi trường Markdown: Gõ lệnh để tạo văn bản giống như đề bài 

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
a = 0; b = 2; c = 4; d = 5
n1 = 2; n2 = 100; n3= 5
x1 = np.linspace(a, b, n1)
y1 = 7*np.ones(n1)
x2 = np.linspace(b, c , n2)
y2 = 2*x2**2 - 5*x2 -2
x3 = np.linspace(c, d , n3)
y3 = -3*x3 +2
x = np.r_[x1, x2, x3];  y = np.r_[y1, y2, y3]
plt.figure(figsize=(6,2))
plt.plot(x, y)
plt.fill_between(x, 0 ,y)
plt.show()
```

## Bài 4. Tính diện tích, chiều dại cạnh, chu vi và trọng tâm của hình phẳng
#### Tọa độ các điểm $[x(i), y(i)]$ của hình phẳng như bảng(thứ tự ngược chiều kim đồng hồ)
|i|0|1|2|3|4|5|
|:--:|-|-|-|-|-|:--:|
|$x_i$|220|499|635|1253|1050|502|
|$y_i$|570|350|94|246|670|855|  


$A = \displaystyle\sum_{i=0}^{n-2}{\dfrac{(x_iy_{i+1}-x_{i+1}y_i)}{2} + \frac{x_{n-1} y_0 - x_0 y_{n-1}}{2}}\\     
L_{i,i < n-1} = \sqrt{(x_{x+1} - x_i)^2 + (y_{i+1}-y_i)^2}\\ 
L_{i, i = n-1} = \sqrt{(x_0 - x_i)^2 + (y_0 - y_i)^2}\\ 
P = \displaystyle\sum_{i=0}^{n-1}{L_i}\\ 
x_c = \displaystyle\sum_{i=0}^{n-1}{\dfrac{(x_i + x_{i+1})(x_i y_{i+1} - x_{i+1} y_i)}{6A}} +  \dfrac{(x_{n-1} + x_0)(x_{n-1} y_0 - x_0 y_{n-1})}{6A}\\ 
y_c = \displaystyle\sum_{i=0}^{n-2}{\dfrac{(y_i + y_{i+1})(x_i y_{i+1} - x_{i+1} y_i)}{6A}} +  \dfrac{(y_{n-1} + y_0)(x_{n-1} y_0 - x_0 y_{n-1})}{6A}
$

```python
%reset -f
import numpy as np
np.set_printoptions(2)
import matplotlib.pyplot as plt
x = np.array([220, 499, 635, 1253, 1050, 502])
y = np.array([570, 350, 94, 246, 670, 855])
n = len(x)
A = 0; P = 0; xc = 0; yc = 0; L = []
for i in range(n):
    if i < n-2:
        A += ((x[i]*y[i+1] - x[i+1]*y[i])/2) 
        Li = np.sqrt((x[i+1]-x[i])**2 + (y[i+1]-y[i])**2)
        L = np.append(L, Li)
        xc += ((x[i]+x[i+1]) * (x[i]*y[i+1] - x[i+1]*y[i]))
        yc += ((y[i]+y[i+1]) * (x[i]*y[i+1] - x[i+1]*y[i]))
    else:
        A += ((x[i]*y[0] - x[0]*y[i])/2)
        Li = np.sqrt((x[0]-x[i])**2 + (y[0]-y[i])**2)
        xc += ((x[i] + x[0]) * (x[i] * y[0] - x[0] * y[i]))
        yc += ((y[i] + y[0]) * (x[i] * y[0] - x[0] * y[i]))
xc /= (6*A)
yc /= (6*A)
A = abs(A)
P = np.sum(L)
print("A=",A)
print("L(i)",L)
print("Xc =", xc)
print("Yc =", yc)
x = np.append(x,x[0]); y = np.append(y,y[0])
plt.plot(x, y, "-bo", label="X và Y")
plt.plot(xc, yc, "rv", label="Xc Và Yc")
plt.title("Hình Phẳng")
plt.legend()
plt.axis("equal")
plt.grid()
plt.show()
```
