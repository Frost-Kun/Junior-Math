```python
#Cell 3
%reset -f
import numpy as np
np.set_printoptions(2, suppress=True)
import matplotlib.pyplot as plt
K = np.zeros((10,10))
f = np.zeros(10); f[2]= -10 ;f[5]= 15
K[0,0]= 1; K[0,3]= 1
K[1,1]= 1; K[1,2]= 1
K[2,4]= 1; K[2,5]= 3/5
K[3,2]= 1; K[3,5]= -4/5
K[4,4]= 1; K[4,7]= 3/5
K[5,6]= 1; K[5,7]= -4/5
K[6,3]= 1; K[6,5]= -3/5
K[7,5]= 4/5; K[7,6]= 1
K[8,7]= -3/5; K[8,8]= -1
K[9,7]= 4/5; K[9,9]= 1
X =np.linalg.solve(K,f)
print("X:= ", X)
print("KT=", K @ X -f)
exs = np.array([[0,3],[3,6],[0,0],
                [0,3],[3,0],[3,3],[6,3]])
eys = np.array([[0,0],[0,0],[0,4],
                [4,0],[4,4],[0,4],[0,4]])
nr = exs.shape[0]
for i in range(nr):
    plt.plot(exs[i], eys[i], "-bo")
plt.grid()
plt.axis("equal")
plt.show()
```
