```python
import numpy as np
a, b, c = 10, 20, 5
s1 = (a + b + c)/3
s2 = a**2 + b**2 + c**3
s3 = np.sqrt(a**2 + b**2) + np.sqrt(b**2 + c**2) + np.sqrt(c**2 + a**2) + np.exp(c)
print("s1=", round(s1,3), "\n s2 =", s2, "\n s3 =", round(s3,2))
```

```python
import numpy as np
a, b, P  = 10, 20 ,50; theta = 30*np.pi/180
N = (P*(a+b))/b*np.sin(theta)
X = N*np.cos(theta)
Y = P - N*np.sin(theta)
R = np.sqrt(X**2 + Y**2)
print("N=", round(N, 2), "\nX =", round(X,2), "\nY =", round(Y,2), "\nR=", round(R,2))
```

```python
import numpy as np
P, Q, a, b, c, b1, h1, b2, h2 = 3., 5., 7., 8., 12., 15., 17., 10., 9.
N1 = (P*a + Q *(a+b))/(a+b+c)
N2 = P + Q - N1
if N1 > abs(N1-P) and N1 > N2:
    Qmax = N1
elif abs(N1-P) > N1 and abs(N1-P) > N2:
        Qmax = abs(N1-P)
else: Qmax = N2
print("Qmax=", round(Qmax,2))
if N1*a > N2*c:
    Mmax = N1*a
else:
    Mmax = N2*c
print("Mmax=", round(Mmax,2))
yc = (0.5*h1*b1*h1 + (h1 + 0.5*h2)*b2*h2)/b1*h1 + b2*h2
Ix = (b1*h1**3)/12 + (0.5*h1 - yc)**2 * b1*h1 + (b2*h2**3)/12 + ( h1 + 0.5*h2 - yc)**2 * b2*h2
Iy = (h1*b1**3)/12 + (h2*b2**3)/12
Sx = 0.5*b1*yc**2
sigmax = (Mmax / Ix) * yc
sigmin = ((-Mmax)/Ix)*(h1+h1-yc)
taumax = (Qmax * Sx) / (Ix*b1)
print("N1=", round(N1,2), "\nN2 =", round(N2,2))
print("yc=", yc, "\nIx=", Ix,"\nIy=", Iy, "\nSx=", Sx)
print("sigmax=", round(sigmax,5), "\nsigmin=", round(sigmin,5))
print("taumax=", round(taumax,2))
```
